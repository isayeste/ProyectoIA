```python
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Cargar el dataframe
df = pd.read_csv("dataset1.csv")

# Seleccionar las columnas num√©ricas
datos_numericos = df.select_dtypes(include=["float64", "int64"])

# Imputar valores faltantes
imputer = SimpleImputer(strategy="median")  # Puedes cambiar la estrategia
datos_imputados = imputer.fit_transform(datos_numericos)

# Normalizar los datos
scaler = StandardScaler()
datos_normalizados = scaler.fit_transform(datos_imputados)

# Aplicar PCA
pca = PCA(n_components=2)
componentes_principales = pca.fit_transform(datos_normalizados)

# Crear un nuevo DataFrame con los resultados
df_pca = pd.DataFrame(componentes_principales, columns=["Componente 1", "Componente 2"])
print(df_pca)
```

```python
import matplotlib.pyplot as plt

# Crear el gr√°fico
plt.figure(figsize=(8, 6))
plt.scatter(df_pca["Componente 1"], df_pca["Componente 2"], alpha=0.6, c="blue", edgecolors="k")
plt.title("Representaci√≥n de PCA", fontsize=14)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(True)
plt.show()
```
